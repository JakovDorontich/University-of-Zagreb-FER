### 1. Linearna regresija kao klasifikator

U prvoj laboratorijskoj vježbi koristili smo model linearne regresije za, naravno, regresiju. Međutim, model linearne regresije može se koristiti i za **klasifikaciju**. Iako zvuči pomalo kontraintuitivno, zapravo je dosta jednostavno. Naime, cilj je naučiti funkciju $f(\mathbf{x})$ koja za negativne primjere predviđa vrijednost $1$, dok za pozitivne primjere predviđa vrijednost $0$. U tom slučaju, funkcija $f(\mathbf{x})=0.5$ predstavlja granicu između klasa, tj. primjeri za koje vrijedi $h(\mathbf{x})\geq 0.5$ klasificiraju se kao pozitivni, dok se ostali klasificiraju kao negativni.

Klasifikacija pomoću linearne regresije implementirana je u razredu [`RidgeClassifier`](http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.RidgeClassifier.html). U sljedećim podzadatcima **istrenirajte** taj model na danim podatcima i **prikažite** dobivenu granicu između klasa. Pritom isključite regularizaciju ($\alpha = 0$, odnosno `alpha=0`). Također i ispišite **točnost** vašeg klasifikacijskog modela (smijete koristiti funkciju [`metrics.accuracy_score`](http://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html)). Skupove podataka vizualizirajte korištenjem pomoćne funkcije ``plot_clf_problem(X, y, h=None)`` koja je dostupna u pomoćnom paketu `mlutils` (datoteku `mlutils.py` možete preuzeti sa stranice kolegija). `X` i `y` predstavljaju ulazne primjere i oznake, dok `h` predstavlja funkciju predikcije modela (npr. `model.predict`). 

U ovom zadatku cilj je razmotriti kako se klasifikacijski model linearne regresije ponaša na linearno odvojim i neodvojivim podatcima.


```python
from sklearn.linear_model import LinearRegression, RidgeClassifier
from sklearn.metrics import accuracy_score
```

#### (a)

Prvo, isprobajte *ugrađeni* model na linearno odvojivom skupu podataka `seven` ($N=7$).

```python
seven_X = np.array([[2,1], [2,3], [1,2], [3,2], [5,2], [5,4], [6,3]])
seven_y = np.array([1, 1, 1, 1, 0, 0, 0])
```

```python
clf = RidgeClassifier(alpha=0.0).fit(seven_X, seven_y)
predicted = clf.predict(seven_X)
print("Dobiveni predviđeni rezultati: "+str(predicted))

plt.figure(figsize=(16,9))
mlutils.plot_2d_clf_problem(seven_X, seven_y, h=clf.predict) 

score = accuracy_score(seven_y, predicted)
print("Dobivena točnost klasifikacijskog modela: "+str(score*100)+" %")
```

### 2. Višeklasna klasifikacija

Postoji više načina kako se binarni klasifikatori mogu se upotrijebiti za višeklasnu klasifikaciju. Najčešće se koristi shema tzv. **jedan-naspram-ostali** (engl. *one-vs-rest*, OVR), u kojoj se trenira po jedan klasifikator $h_j$ za svaku od $K$ klasa. Svaki klasifikator $h_j$ trenira se da razdvaja primjere klase $j$ od primjera svih drugih klasa, a primjer se klasificira u klasu $j$ za koju je $h_j(\mathbf{x})$ maksimalan.

Pomoću funkcije [`datasets.make_classification`](http://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_classification.html) generirajte slučajan dvodimenzijski skup podataka od tri klase i prikažite ga koristeći funkciju `plot_2d_clf_problem`. Radi jednostavnosti, pretpostavite da nema redundantnih značajki te da je svaka od klasa "zbijena" upravo u jednu grupu.

```python
from sklearn.datasets import make_classification
```

```python
n = 100
X, y = make_classification(n_samples=n, n_features=2, n_redundant=0, n_classes=3, n_clusters_per_class=1, shuffle = True)

plt.figure(figsize=(16,9))
mlutils.plot_2d_clf_problem(X, y, h=None)


y0 = np.array([])
y1 = np.array([])
y2 = np.array([])

for i in range(0, n):
    if y[i]==0:
        y0 = np.append(y0, 1)
        y1 = np.append(y1, 0)
        y2 = np.append(y2, 0)
    if y[i]==1:
        y0 = np.append(y0, 0)
        y1 = np.append(y1, 1)
        y2 = np.append(y2, 0)
    if y[i]==2:
        y0 = np.append(y0, 0)
        y1 = np.append(y1, 0)
        y2 = np.append(y2, 1)

```

Trenirajte tri binarna klasifikatora, $h_1$, $h_2$ i $h_3$ te prikažite granice između klasa (tri grafikona). Zatim definirajte $h(\mathbf{x})=\mathrm{argmax}_j h_j(\mathbf{x})$  (napišite svoju funkciju `predict` koja to radi) i prikažite granice između klasa za taj model. Zatim se uvjerite da biste identičan rezultat dobili izravno primjenom modela `RidgeClassifier`, budući da taj model za višeklasan problem zapravo interno implementira shemu jedan-naspram-ostali.

**Q:** Alternativna shema jest ona zvana **jedan-naspram-jedan** (engl, *one-vs-one*, OVO). Koja je prednost sheme OVR nad shemom OVO? A obratno?

```python
plt.figure(figsize=(24,5))
clf0 = LinearRegression().fit(X, y0)
h1 = lambda x : clf0.predict(x) >= 0.5        
plt.subplot(1,3,1).set_title("Klasa 1")
mlutils.plot_2d_clf_problem(X, y0, h=h1)

clf1 = LinearRegression().fit(X, y1)
h2 = lambda x : clf1.predict(x) >= 0.5     
plt.subplot(1,3,2).set_title("Klasa 2")
mlutils.plot_2d_clf_problem(X, y1, h=h2)

clf2 = LinearRegression().fit(X, y2)
h3 = lambda x : clf2.predict(x) >= 0.5      
plt.subplot(1,3,3).set_title("Klasa 3")
mlutils.plot_2d_clf_problem(X, y2, h=h3)
plt.show()

def predict(X):
    yh1 = clf0.predict(X)
    yh2 = clf1.predict(X)
    yh3 = clf2.predict(X)
    out = np.array([])
    for i in range(0, len(yh1)):
        val = [yh1[i], yh2[i], yh3[i]]
        out = np.append(out, argmax(val))
    return out

plt.figure(figsize=(16,9))
h_all = lambda x : predict(x)
mlutils.plot_2d_clf_problem(X, y, h=h_all)
plt.show()


clf = RidgeClassifier().fit(X, y)
plt.figure(figsize=(16,9))
mlutils.plot_2d_clf_problem(X, y, h=clf.predict)
plt.show()



predicted = predict(X)
print("OVR rezultati ručno:")
print(predicted)

predicted2 = clf.predict(X)
print("OVR rezultati automatski:")
print(predicted2)

flag = True
for i in range(0, len(predicted)):
    if predicted[i] != predicted2[i]:
        flag = False
if flag == True:
    print("Rezultati su identični!")
else:
    print("Rezultati jesu različiti!")
    


```

### 3. Logistička regresija

Ovaj zadatak bavi se probabilističkim diskriminativnim modelom, **logističkom regresijom**, koja je, unatoč nazivu, klasifikacijski model.

Logistička regresija tipičan je predstavnik tzv. **poopćenih linearnih modela** koji su oblika: $h(\mathbf{x})=f(\mathbf{w}^\intercal\tilde{\mathbf{x}})$. Logistička funkcija za funkciju $f$ koristi tzv. **logističku** (sigmoidalnu) funkciju $\sigma (x) = \frac{1}{1 + \textit{exp}(-x)}$.

#### (a)

Definirajte logističku (sigmoidalnu) funkciju $\mathrm{sigm}(x)=\frac{1}{1+\exp(-\alpha x)}$ i prikažite je za $\alpha\in\{1,2,4\}$.

```python
def sigm(x, alpha):
    temp = 1 + exp(-1 * alpha * x)
    return 1/temp

a = [1,2,4]
X = range(-6,6, 1)
plt.figure(figsize=(16,9))
for alpha in a:
    Y = []
    for x in X:
        y = sigm(x, alpha)
        Y.append(y)
    
    plt.plot(X,Y, label='alpha:'+str(alpha))
plt.legend(loc="center right")


        
```

**Q**: Zašto je sigmoidalna funkcija prikladan izbor za aktivacijsku funkciju poopćenoga linearnog modela? 
</br>

**Q**: Kakav utjecaj ima faktor $\alpha$ na oblik sigmoide? Što to znači za model logističke regresije (tj. kako izlaz modela ovisi o normi vektora težina $\mathbf{w}$)?

#### (b)

Implementirajte funkciju 

> `lr_train(X, y, eta=0.01, max_iter=2000, alpha=0, epsilon=0.0001, trace=False)` 

za treniranje modela logističke regresije gradijentnim spustom (*batch* izvedba). Funkcija uzima označeni skup primjera za učenje (matrica primjera `X` i vektor oznaka `y`) te vraća $(n+1)$-dimenzijski vektor težina tipa `ndarray`. Ako je `trace=True`, funkcija dodatno vraća listu (ili matricu) vektora težina $\mathbf{w}^0,\mathbf{w}^1,\dots,\mathbf{w}^k$ generiranih kroz sve iteracije optimizacije, od 0 do $k$. Optimizaciju treba provoditi dok se ne dosegne `max_iter` iteracija, ili kada razlika u pogrešci unakrsne entropije između dviju iteracija padne ispod vrijednosti `epsilon`. Parametar `alpha` predstavlja faktor L2-regularizacije.

Preporučamo definiranje pomoćne funkcije `lr_h(x,w)` koja daje predikciju za primjer `x` uz zadane težine `w`. Također, preporučamo i funkciju `cross_entropy_error(X,y,w)` koja izračunava pogrešku unakrsne entropije modela na označenom skupu `(X,y)` uz te iste težine.

**NB:** Obratite pozornost na to da je način kako su definirane oznake ($\{+1,-1\}$ ili $\{1,0\}$) kompatibilan s izračunom funkcije gubitka u optimizacijskome algoritmu.

```python
from numpy import linalg
from sklearn.preprocessing import PolynomialFeatures

def pred(X, w):
    poly = PolynomialFeatures(1)
    fiX = poly.fit_transform(X)
    temp = np.dot(fiX,w)
    return sigm(temp, 1)

def lr_h(x, w):
    wt = np.transpose(w)
    temp = np.matmul(wt, x)
    return sigm(temp, 1)

def cross_entropy_error(X, y, w): 
    N = X.shape[0]
    suma = 0
    for i in range(N):
        h = lr_h(X[i], w)
        suma += -1 * y[i] * math.log(h) - (1 - y[i]) * math.log(1 - h)
    return (suma/N)

def lr_train(X, y, eta=0.01, max_iter=2000, trace=False, epsilon=0.000001):
    N = X.shape[0]
    n = X.shape[1]
    w = np.array([])
    iter = 0
    for i in range(0, n):
        w = np.append(w, 0)
    derr = 0
    lasterr = 0
    currerr = 0
    W = np.array([w])
    while(iter < max_iter):
        dw = np.array([])
        for i in range(0, n):
            dw = np.append(dw, 0)
        
        for i in range(N):
            h= lr_h(X[i], w)
            dw = dw - ((h-y[i])*X[i])
        w = w + eta*dw
        W = np.vstack((W,w))
        currerr = cross_entropy_error(X, y, w)
        if(abs(currerr - lasterr) <= epsilon):  
            if(trace == True):
                return tezine
            else:
                return w
        
        lasterr = currerr
        iter += 1
       
    if(trace == True):
        
        return W
    else:
        return w


```

#### (c)

Koristeći funkciju `lr_train`, trenirajte model logističke regresije na skupu `seven`, prikažite dobivenu granicu između klasa te  izračunajte pogrešku unakrsne entropije. 

**NB:** Pripazite da modelu date dovoljan broj iteracija.

```python
poly = PolynomialFeatures(1)
fiX = poly.fit_transform(seven_X)
w = lr_train(fiX, seven_y, trace = False)

print("Dobivene težine modela: "+str(w))

predict = lambda x : pred(x,w) >= 0.5

predicted = []
for p in pred(seven_X, w):
    if p >= 0.5:
        predicted.append(1.0)
    else:
        predicted.append(0.0)
print("Predviđene vrijednosti:"+str(predicted))

pogreska = cross_entropy_error(fiX,seven_y,w)
print("Pogreška unakrsne entropije: "+str(pogreska))

score = accuracy_score(seven_y, predicted)
print("Dobivena točnost klasifikacijskog modela: "+str(score*100)+" %")

plt.figure(figsize=(16,9))
mlutils.plot_2d_clf_problem(seven_X, seven_y, h=predict) 
```

### 5. Regularizirana logistička regresija

Trenirajte model logističke regresije na skupu `seven` s različitim faktorima L2-regularizacije, $\alpha\in\{0,1,10,100\}$. Prikažite na dva odvojena grafikona (1) pogrešku unakrsne entropije te (2) L2-normu vektora $\mathbf{w}$ kroz iteracije optimizacijskog algoritma.

**Q:** Jesu li izgledi krivulja očekivani i zašto?

**Q:** Koju biste vrijednost za $\alpha$ odabrali i zašto?

```python
from numpy.linalg import norm
```

```python
def lr_train_reg(X, y, eta=0.01, max_iter=2000, trace=False, alpha=0, epsilon=0.000001):
    N = X.shape[0]
    n = X.shape[1]
    iter = 0
    w = np.array([])
    for i in range(0, n-1):
        w = np.append(w, 0)
    w0 = 0
    W=np.array([w0])
    W=np.append(W,w)
    lasterr = 0
    currerr = 0
    tezine = np.array([W])
    while(iter < max_iter):
        dw0 = 0
        dw = np.array([])
        for i in range(0, n-1):
            dw = np.append(dw, 0)
        
        for i in range(N):
            h= lr_h(X[i], W)
            dw0 = dw0 - (h - y[i])
            grad = ((h-y[i])*X[i])
            dw = dw - grad[1:3]
        w0 = w0 + eta*dw0
        w = w*(1-eta*alpha) + eta*dw
        W=np.array([w0])
        W=np.append(W,w) 
       
        tezine = np.vstack((tezine,W))
        currerr = cross_entropy_error(X, y, W)
        if(abs(currerr - lasterr) <= epsilon):  
            if(trace == True):
                return tezine
            else:
                return w
        
        lasterr = currerr
        iter += 1
       
    if(trace == True):
        return tezine
    else:
        return W
    
```

```python
alpha = [0, 1, 10, 100]
poly = PolynomialFeatures(1)
fiX = poly.fit_transform(seven_X)
plt.figure(figsize=(16,9))
plt.xlabel('Iteracija')
plt.ylabel('Pogreška unakrsne entropije')
colors = ['yellow','green','blue','red']
for i in range(0,4):
    tezine = lr_train_reg(fiX, seven_y,alpha=alpha[i], trace = True)
    iteracije = range(0,len(tezine))
    PUE = np.array([])
    for w in tezine:
        pogreska_entropije = cross_entropy_error(fiX, seven_y, w)
        PUE = np.append(PUE,pogreska_entropije)
    plt.plot(iteracije, PUE , colors[i], label="Alpha ="+str(alpha[i]))
plt.legend(loc="center right")
plt.show()

plt.figure(figsize=(16,9))
plt.xlabel('Iteracija')
plt.ylabel('L2 Norma težina')
for i in range(0,4):
    tezine = lr_train_reg(fiX, seven_y,alpha=alpha[i], trace = True)
    iteracije = range(0,len(tezine))
    norme = np.array([])
    for w in tezine:
        L2 = norm(w[1:len(w)])
        norme = np.append(norme,L2)
    plt.plot(iteracije, norme , colors[i], label="Alpha ="+str(alpha[i]))
    

plt.legend(loc="center right")
plt.show()
```
